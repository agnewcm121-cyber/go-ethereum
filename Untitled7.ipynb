```python
#!/usr/bin/env python3
"""
Great-circle "between" finder for MrBeast puzzle:
- Stage 1 = 51 endpoint cities (embedded below)
- Stage 2 = 40 midpoint cities (embedded below)

It geocodes both lists (once), caches to CSV, then:
For every Stage1 pair (A,B) it finds Stage2 cities P that are:
  - near the great-circle A->B (cross-track <= threshold)
  - AND between endpoints along-track (0..total distance, with optional endpoint buffer)

Outputs:
  stage1_endpoints_geocoded.csv
  stage2_midpoints_geocoded.csv
  matches.csv
"""

from __future__ import annotations
import csv
import math
import time
import itertools
from dataclasses import dataclass
from typing import List, Tuple, Iterable, Optional

EARTH_RADIUS_KM = 6371.0088


# ----------------------------
# ✅ Stage 1 endpoint cities (51) — from your screenshots
# ----------------------------
STAGE1_CITIES = [
    "Maputo, Mozambique",
    "Tashkent, Uzbekistan",
    "Arles, France",
    "Kabul, Afghanistan",
    "Accra, Ghana",
    "Lahore, Pakistan",
    "Tallinn, Estonia",
    "New Orleans, Louisiana",
    "Lima, Peru",
    "Casablanca, Morocco",
    "Kupang, Indonesia",
    "Yellowstone",
    "North York, Ontario",
    "Kandi, Benin",
    "Seoul, South Korea",
    "Cairo, Egypt",
    "Lower Hutt, New Zealand",
    "Ankara, Turkey",
    "Tbilisi, Georgia",
    "Toad Suck, Arkansas",
    "Tijuana, Mexico",
    "Armavir, Armenia",
    "Salina Turda, Romania",
    "Doha, Qatar",
    "Moscow, Russia",
    "Algiers, Algeria",
    "Tampa Bay, Florida",
    "Wichita, Kansas",
    "Tierra Del Fuego",
    "Dekalb, Illinois",
    "Montreal, Quebec",
    "Queensland, Australia",
    "Divo, Cote d'Ivoire",
    "Copenhagen, Denmark",
    "Buffalo, New York",
    "Sucre, Bolivia",
    "Male, Maldives",
    "Thane, India",
    "Iturup, Russia",
    "Yellowknife, Northwest Territories",
    "Nome, Alaska",
    "Dimtu, Ethiopia",
    "Silver Coast, Portugal",
    "Antananarivo, Madagascar",
    "Gaborone, Botswana",
    "Quito, Ecuador",
    "South Pole",
    "Heard Island",
    "Ascension Island",
    "Ile de la Possession",
    "Eyjafjallajokull, Iceland",
]


# ----------------------------
# ✅ Stage 2 midpoint cities (40) — from your typed list
# ----------------------------
STAGE2_CITIES = [
    "Casper, Wyoming",
    "Arzamas, Russia",
    "Mubi, Nigeria",
    "Baku, Azerbaijan",
    "Moora, Australia",
    "Surat, India",
    "Marquette, Michigan",
    "Beatty, Nevada",
    "Mosul, Iraq",
    "Visby, Sweden",
    "Istanbul, Turkey",
    "Genoa, Italy",
    "Mayotte",
    "Giresun, Turkey",
    "L'Ascension, Quebec",
    "Okato, New Zealand",
    "Nova Russas, Ceara, Brazil",
    "Ceske Budejovice, Czechia",
    "Christchurch, New Zealand",
    "Castelo Branco, Portugal",
    "Sokode, Togo",
    "Curico, Chile",
    "Monkey Bay, Malawi",
    "Antalya, Turkey",
    "Ambanja, Madagascar",
    "Saint-Pierre",
    "Krasnodar, Russia",
    "San Miguel de Tucuman, Argentina",
    "Derby Wharf, Australia",
    "Manaus, Brazil",
    "Pune, India",
    "Natif Waterfalls, Oman",
    "Velingrad, Bulgaria",
    "Marrakesh, Morocco",
    "Orlando, Florida",
    "Charlotte, North Carolina",
    "Edmonton, Alberta",
    "Vladivostok, Russia",
    "Cochabamba, Bolivia",
    "Tacurong, CSK, Philippines",
]


@dataclass(frozen=True)
class City:
    name: str
    lat: float
    lon: float


# ---------- geo helpers ----------

def deg2rad(d: float) -> float:
    return d * math.pi / 180.0

def wrap_lon_deg(lon: float) -> float:
    return (lon + 180.0) % 360.0 - 180.0

def haversine_km(lat1: float, lon1: float, lat2: float, lon2: float) -> float:
    φ1, λ1, φ2, λ2 = map(deg2rad, [lat1, lon1, lat2, lon2])
    dφ = φ2 - φ1
    dλ = λ2 - λ1
    a = math.sin(dφ/2)**2 + math.cos(φ1)*math.cos(φ2)*math.sin(dλ/2)**2
    c = 2 * math.asin(min(1.0, math.sqrt(a)))
    return EARTH_RADIUS_KM * c

def initial_bearing_rad(lat1: float, lon1: float, lat2: float, lon2: float) -> float:
    φ1, λ1, φ2, λ2 = map(deg2rad, [lat1, lon1, lat2, lon2])
    dλ = λ2 - λ1
    y = math.sin(dλ) * math.cos(φ2)
    x = math.cos(φ1)*math.sin(φ2) - math.sin(φ1)*math.cos(φ2)*math.cos(dλ)
    return math.atan2(y, x)

def cross_track_and_along_track_km(a: City, b: City, p: City) -> Tuple[float, float, float]:
    d13_km = haversine_km(a.lat, a.lon, p.lat, p.lon)
    d12_km = haversine_km(a.lat, a.lon, b.lat, b.lon)
    if d12_km == 0:
        return (float("inf"), float("inf"), 0.0)

    θ13 = initial_bearing_rad(a.lat, a.lon, p.lat, p.lon)
    θ12 = initial_bearing_rad(a.lat, a.lon, b.lat, b.lon)

    δ13 = d13_km / EARTH_RADIUS_KM

    # cross-track
    δxt = math.asin(max(-1.0, min(1.0, math.sin(δ13) * math.sin(θ13 - θ12))))
    cross_track_km = δxt * EARTH_RADIUS_KM

    # along-track
    δat = math.atan2(
        math.sin(δ13) * math.cos(θ13 - θ12),
        max(1e-15, math.cos(δ13))
    )
    along_track_km = δat * EARTH_RADIUS_KM

    return (cross_track_km, along_track_km, d12_km)


# ---------- CSV cache helpers ----------

def read_cities_csv(path: str) -> List[City]:
    out: List[City] = []
    with open(path, "r", newline="", encoding="utf-8") as f:
        reader = csv.DictReader(f)
        for row in reader:
            out.append(City(
                name=row["name"].strip(),
                lat=float(row["lat"]),
                lon=wrap_lon_deg(float(row["lon"]))
            ))
    return out

def write_cities_csv(path: str, cities: List[City]) -> None:
    with open(path, "w", newline="", encoding="utf-8") as f:
        w = csv.DictWriter(f, fieldnames=["name", "lat", "lon"])
        w.writeheader()
        for c in cities:
            w.writerow({"name": c.name, "lat": c.lat, "lon": c.lon})

def write_matches_csv(path: str, rows: List[dict]) -> None:
    if not rows:
        print("No matches found.")
        return
    with open(path, "w", newline="", encoding="utf-8") as f:
        w = csv.DictWriter(f, fieldnames=list(rows[0].keys()))
        w.writeheader()
        w.writerows(rows)


# ---------- geocode + cache ----------

def geocode_list(names: List[str], cache_path: str, user_agent: str) -> List[City]:
    """
    Geocode city names using Nominatim and cache to CSV.
    Re-geocodes only missing items.
    """
    try:
        from geopy.geocoders import Nominatim
    except ImportError:
        raise SystemExit(
            "Missing dependency: geopy\n"
            "Install with: pip install geopy\n"
            "Or manually create the cache CSV with columns: name,lat,lon."
        )

    cached: dict[str, City] = {}
    try:
        existing = read_cities_csv(cache_path)
        cached = {c.name: c for c in existing}
        print(f"Loaded {len(cached)} cached from {cache_path}")
    except FileNotFoundError:
        pass

    geolocator = Nominatim(user_agent=user_agent)

    out: List[City] = []
    for name in names:
        if name in cached:
            out.append(cached[name])
            continue

        loc = None
        for attempt in range(3):
            try:
                loc = geolocator.geocode(name, timeout=20)
                break
            except Exception:
                time.sleep(2.0 * (attempt + 1))

        if not loc:
            print(f"[WARN] Could not geocode: {name}")
            continue

        city = City(name=name, lat=float(loc.latitude), lon=wrap_lon_deg(float(loc.longitude)))
        out.append(city)
        print(f"Geocoded: {name} -> ({city.lat:.5f}, {city.lon:.5f})")
        time.sleep(1.1)  # be polite to Nominatim

    write_cities_csv(cache_path, out)
    print(f"Wrote {len(out)} rows to {cache_path}")
    return out


# ---------- between logic ----------

def find_between_for_pair(
    a: City,
    b: City,
    candidates: Iterable[City],
    cross_track_km: float,
    endpoint_buffer_km: float,
) -> List[Tuple[City, float, float, float]]:
    hits = []
    for p in candidates:
        xt_km, at_km, total_km = cross_track_and_along_track_km(a, b, p)

        # Between endpoints along the track
        if at_km < endpoint_buffer_km:
            continue
        if at_km > total_km - endpoint_buffer_km:
            continue

        # Near the great-circle
        if abs(xt_km) <= cross_track_km:
            hits.append((p, abs(xt_km), at_km, total_km))

    hits.sort(key=lambda t: (t[1], t[2]))
    return hits


def main():
    # ---- knobs to tune ----
    CROSS_TRACK_KM = 150.0       # try 50, 100, 150, 200 depending how strict you want
    ENDPOINT_BUFFER_KM = 50.0    # avoid candidates that are basically at endpoints
    # -----------------------

    stage1 = geocode_list(STAGE1_CITIES, "stage1_endpoints_geocoded.csv", "stage1_geocoder")
    stage2 = geocode_list(STAGE2_CITIES, "stage2_midpoints_geocoded.csv", "stage2_geocoder")

    rows: List[dict] = []
    for a, b in itertools.combinations(stage1, 2):
        hits = find_between_for_pair(a, b, stage2, CROSS_TRACK_KM, ENDPOINT_BUFFER_KM)
        for p, xt, at, total in hits:
            rows.append({
                "endpoint_a": a.name,
                "endpoint_b": b.name,
                "stage2_midpoint": p.name,
                "cross_track_km": round(xt, 3),
                "along_track_km": round(at, 1),
                "total_km": round(total, 1),
            })

    rows.sort(key=lambda r: (r["cross_track_km"], r["total_km"]))
    write_matches_csv("matches.csv", rows)
    print(f"Done. Wrote {len(rows)} rows to matches.csv")


if __name__ == "__main__":
    main()
```
